```python
import pandas as pd
df=pd.read_csv("./Combined_News_DJIA.csv")
```

```python
df.sample(10)
```

```python
df.info()
```

```python
train=df[df["Date"]<'20150101']
test=df[df["Date"]>'20141231']
```

```python
train
```

```python
data=train.iloc[:,2:27]
data.replace("[^a-zA-Z]"," ",regex=True,inplace=True)
```

```python
list2=[i for i in range(25)]
new_index=[str(i)  for i in list2]
data.columns=new_index
data.head()
```

```python
for index in new_index:
  data[index]=data[index].str.lower()
data.head()
```

```python
def remove_b(cell):
    if isinstance(cell, str) and cell.startswith('b'):
        return cell[1:]
    else:
        return cell
data = data.applymap(remove_b)
```

```python
data.head()
```

```python
" ".join(str(x) for x in data.iloc[1,0:25])
```

```python
headline=[]
for row in range(0,len(data.index)):
  headline.append(" ".join(str(x) for x in data.iloc[row,0:25]))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.ensemble import RandomForestClassifier
cv=CountVectorizer(ngram_range=(2,2))
traindata=cv.fit_transform(headline)
rf=RandomForestClassifier(n_estimators=200,criterion="entropy")
rf.fit(traindata,train["Label"])
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import GradientBoostingClassifier
cv=TfidfVectorizer(ngram_range=(2,2))
traindata=cv.fit_transform(headline)
rf=GradientBoostingClassifier(n_estimators=100, learning_rate=0.1,max_depth=8, random_state=0)
rf.fit(traindata,train["Label"])
```

```python
test_trans=[]
for row in range(0,len(test.index)):
  test_trans.append(" ".join(str(x) for x in test.iloc[row,2:27]))
test_data=cv.transform(test_trans)
pred=rf.predict(test_data)
```

```python
pred
```

```python
from sklearn.metrics import classification_report,accuracy_score
accuracy_score(test['Label'],pred)
```

```python
print(classification_report(test['Label'],pred))
```
